```python
import pandas as pd
benchmarks1 = pd.read_csv("benchmarks_v01.tsv", sep="\t")
benchmarks1.head()
benchmarks2 = pd.read_csv("benchmarks_v02.tsv", sep="\t")
benchmarks2.head()
```

```python
import altair as alt
alt.renderers.enable('default')

chart1 = alt.Chart(benchmarks1).mark_bar().encode(
    alt.X('nsim:N'),
    alt.Y('time', scale=alt.Scale(type='linear')),
    color='key',
    column='time_type'
    
).interactive()

chart2 = alt.Chart(benchmarks2).mark_bar().encode(
    alt.X('nsim:N'),
    alt.Y('time', scale=alt.Scale(type='linear')),
    color='key',
    column='time_type'
    
).interactive()


chart1 & chart2
```

```python
chart1 = alt.Chart(benchmarks1).mark_point().encode(
    alt.X('nsim:Q', scale=alt.Scale(type="log")),
    alt.Y('time', scale=alt.Scale(type="log")),
    color='key',
    tooltip="time_type",
    shape="time_type"
    
).interactive()

chart2 = alt.Chart(benchmarks2).mark_point().encode(
    alt.X('nsim:Q', scale=alt.Scale(type="log")),
    alt.Y('time', scale=alt.Scale(type="log")),
    color='key',
    tooltip="time_type",
    shape="time_type"
    
).interactive()

chart1 & chart2
```
